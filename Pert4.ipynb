# Konversi data tersebut menjadi tabulasi

```python
col = set()
for trc in dataset:
  for item in trc:
    col.add(item)

col = set(col)
col
```

```python
print(col) 
len(col)
```

```python
data = []
for data in dataset:
  kolom_tabulasi = []
  for item in col:
    for item_data in data:
      biner = 0
      if item_data == item:
        biner = 1
        kolom_tabulasi.append(biner)

  data.append(kolom_tabulasi)

data
```
